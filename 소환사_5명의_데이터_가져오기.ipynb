```
API = "RGAPI-d026e00a-7742-4520-b216-adc3ae4a35e1"   # 라이엇 api 키
df = pd.DataFrame(columns = col)   # 최종으로 사용할 데이터프레임

name0, name1, name2, name3, name4 = input("소환사 5명의 이름을 입력하세요 : ").split(",")
names = [name0, name1, name2, name3, name4]

for i in range(5):
  dff = pd.DataFrame(columns = col)
  name = names[i]
  name = name.replace(" ", "")     

  URL1 =  "https://kr.api.riotgames.com/lol/summoner/v4/summoners/by-name/" + urllib.parse.quote_plus(name, encoding='utf-8') + "?api_key=" + API
  try:
    r = requests.get(URL1, timeout =30)
    while r.status_code!=200: 
      time.sleep(5)
      r = requests.get(URL1)                  
  except Exception as e:
    print(e)

  basic = json.loads(r.text)   

  accountId = basic['accountId']

  URL2 = "https://kr.api.riotgames.com/lol/match/v4/matchlists/by-account/" + accountId + "?api_key=" + API

  try:
    r = requests.get(URL2, timeout =30)
    while r.status_code!=200:
      time.sleep(5)
      r = requests.get(URL2)                  
  except Exception as e:
    print(e)
  data = json.loads(r.text)

  # 최근 100게임 중 랭크 게임의 게임 id 가져오기 
  gameid = []
  k = 0

  while k < 100:
    if data["matches"][k]["queue"] == 420:
      gameid.append(data["matches"][k]["gameId"])
      k = k + 1
    else:
      k = k + 1
      pass

  for i in range(len(gameid)):
    URL3 = "https://kr.api.riotgames.com/lol/match/v4/matches/" + str(gameid[i]) + "?api_key=" + API
    try:
      r = requests.get(URL3, timeout =30)
      while r.status_code!=200: # 요청 제한 또는 오류로 인해 정상적으로 받아오지 않는 상태라면, 5초 간 시간을 지연
        time.sleep(5)
        r = requests.get(URL3)         
    except Exception as e:
      print(e)
    stat = json.loads(r.text)

    for n in range(1, 10):
      if stat["participantIdentities"][n]["player"]["summonerName"] == name:
        participantId = stat["participantIdentities"][n]["participantId"]
      else:
        pass

    for w in range(10):
      if stat["participants"][w]["participantId"] == participantId:
        stats = stat["participants"][w]["stats"]
      else:
        pass 

    
    a = pd.DataFrame(stats, columns = col, index = [0])
    dff = pd.concat([dff, a], axis = 0)

  df_new = pd.DataFrame(np.mean(dff))
  df_new = df_new.T
  df = pd.concat([df, df_new], axis = 0)

```
