```python
# 패키지 불러오기
import requests # api 요청
import json
import pandas as pd
import numpy as np
import time
import urllib.parse
```

```python
# 라이엇에서 기본 정보 가져오기

name = input("소환사 이름 : ")                                                           #찾고자 하는 소환사 이름
name = name.replace(" ", "")                                                            #소환사 이름에 공백이 있다면 제거
API = ""   #라이엇 api 키

URL =  "https://kr.api.riotgames.com/lol/summoner/v4/summoners/by-name/" + urllib.parse.quote_plus(name, encoding='utf-8') + "?api_key=" + API
r = requests.get(URL)         
basic = json.loads(r.text)   

id = basic['id']
accountId = basic['accountId']
puuid = basic['puuid']

print("id : ", id); print("accountId : ", accountId); print("puuid : ", puuid);
```

```python
# 소환사 랭크 정보 가져오기
URL = "https://kr.api.riotgames.com/lol/league/v4/entries/by-summoner/" + id + "?api_key=" + API
r = requests.get(URL)   
rank = json.loads(r.text)    

leagueId = rank[0]['leagueId']
summonerId = rank[0]['summonerId']
tier = rank[0]['tier']
num = rank[0]['rank']

print("tier : ", tier, " ", num)
print("leagueId : ", leagueId); print("summonerId : ", summonerId)
```

```python
# 같은 리그에 속한 소환사들 정보 가져오기
URL = "https://kr.api.riotgames.com/lol/league/v4/leagues/" + leagueId + "?api_key=" + API
r = requests.get(URL)
league = json.loads(r.text)

num = 0
league_list = []
for i in range(len(league['entries'])):
    league_list.append(league['entries'][i]['summonerName'])
    num = num + 1
print("리그에 속한 소환사 : " + str(num) + " 명")
print(league_list)
```
